# Passing solution using two maps, left and right

```python
#!/bin/python3

import math
import os
import random
import re
import sys
from collections import defaultdict

# Complete the countTriplets function below.
def countTriplets(arr, r):
    n = len(arr)
    
    left_map = defaultdict(int)
    right_map = defaultdict(int)
    
    for i in range(n):
        right_map[arr[i]] += 1
    
    count = 0
    
    for i in range(n):
        aj = arr[i]
        right_map[aj] -= 1
        
        aj_l = aj // r
        aj_r = aj * r
        #if left_map[aj_l] > 0 and right_map[aj_r] > 0:
        ctl = left_map[aj_l]
        ctr = right_map[aj_r]
        
        if aj % r == 0:
            count += ctl * ctr
        
        left_map[aj] += 1
                
    
    
    return count
            

def count_less(arr, thr, n):
    i = 0
    
    while i < n and arr[i] < thr:
        i += 1
    
    return max(i - 1, 0)

    
    
    
    
def count_triplets(arr, r, i):
    first = arr[i]
    n = len(arr)
    second = first * r
    third = second * r
    ct = 0
    si = i + 1
    while si < n:
        if arr[si] == second:
            ti = si + 1
            while ti < n:
                if arr[ti] == third:
                    ct += 1
                ti += 1
        si += 1
    
    return ct
if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    nr = input().rstrip().split()

    n = int(nr[0])

    r = int(nr[1])

    arr = list(map(int, input().rstrip().split()))

    ans = countTriplets(arr, r)

    fptr.write(str(ans) + '\n')

    fptr.close()
```
