```python
def flip_vertical(maty):
    matrix2 = list(maty)
    m = len(matrix2)
    n = len(matrix2[0])
    
    i = 0
    j = n-1
    
    while i < j:
        swap_cols(i,j,matrix2)
        i += 1
        j -= 1
    return matrix2

def swap_rows(i,j,matrix):
    print(id(matrix))
    n = len(matrix[0])
    
    for k in range(n):
        b_el = matrix[j][k]
        matrix[j][k] = matrix[i][k]
        matrix[i][k] = b_el
    
def swap_cols(i,j,matrix):
    
    m = len(matrix)
    
    for k in range(m):
        r_el = matrix[k][j]
        matrix[k][j] = matrix[k][i]
        matrix[k][i] = r_el
        
def sum_columns(matrix):
    
    m = len(matrix)
    n = len(matrix[0])
    
    col_sums = []
    
    
    for i in range(n):
        tot = 0
        for k in range(m):
            
            tot += matrix[k][i]
        
        col_sums.append(tot)
    
    return col_sums


def sum_rows(matrix):
    
    m = len(matrix)
    n = len(matrix[0])
    row_sums = []
    
    for i in range(m):
        tot = 0
        for j in range(n):
            tot += matrix[i][j]
        
        row_sums.append(tot)
    return row_sums
```
