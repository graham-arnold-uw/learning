```python
def solve(b, i, j, currTry):
    if currTry > maxTry:
        return b
    if b.isSolved():
        return b
    
    dirs = b.validDirs(i,j)
    
    ctry = b
    
    #up try
    if dirs[0] == 1:
        k = i-1
        m = j
        b.swap(i,j,k,m)
        currTry += 1
        ctry = solve(b,k,m, currTry)
        if uptry.isSolved():
            return ctry
        currTry -= 1
        b.swap(i,j,k,m)
        
    #down try
    if dirs[1] == 1:
        k = i+1
        m = j
        b.swap(i,j,k,m)
        currTry += 1
        ctry = solve(b,k,m, currTry)
        if uptry.isSolved():
            return ctry
        currTry -= 1
        b.swap(i,j,k,m)
        
    #left try
    if dirs[2] == 1:
        k = i
        m = j-1
        b.swap(i,j,k,m)
        currTry += 1
        ctry = solve(b,k,m,m currTry)
        if uptry.isSolved():
            return ctry
        currTry -= 1
        b.swap(i,j,k,m)
        
    #right try
    if dirs[3] == 1:
        k = i
        m = j+1
        b.swap(i,j,k,m)
        currTry += 1
        ctry = solve(b,k,m, currTry)
        if uptry.isSolved():
            return ctry
        currTry -= 1
        b.swap(i,j,k,m)
    
    return cTry
            
    

def solveBoard(inputBoard, i, j):
    
    b = Board(inputBoard)
    maxTry = 100
    currTry = 0
    
    res = solve(b,i,j, currTry, maxTry)
    
    
    if res.isSolved() == 0:
        print("no result found")
    else:
        print res.board
    
```
