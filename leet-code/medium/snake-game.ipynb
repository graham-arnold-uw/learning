```python
from collections import deque

class SnakeGame:

    def __init__(self, width: int, height: int, food: List[List[int]]):
        self.length = 0
        self.food = food
        self.num_food = len(self.food)
        self.width = width
        self.height = height
        #self.q = deque()
        #self.q.append((0,0))
        #self.food_idx = 0
        self.occupied = [[0 for j in range(width)] for i in range(height)]
        #self.occupied[0][0] = 1
        self.dir_map = {'U': (-1, 0), 'D': (1, 0), 'L': (0, -1), 'R': (0, 1)}
        self.head = (0,0)
        self.tail = (0,0)
        
    def move(self, direction: str) -> int:
        hx, hy = self.head
        dx, dy = self.dir_map[direction]
        
        nx, ny = hx + dx, hy + dy
        
        #if nx or ny out of bounds -> Game over
        if nx < 0 or nx >= self.height or ny < 0 or ny >= self.width:
            return -1
        
        elif self.length < self.num_food and nx == self.food[self.length][0] and ny == self.food[self.length][1]:
            self.occupied[hx][hy] = (nx,ny)
            self.head = (nx,ny)
            self.length += 1
        
        elif self.occupied[nx][ny] == 0:
            self.occupied[hx][hy] = (nx, ny)
            self.head = (nx, ny)
            tx,ty = self.tail
            ntx, nty = self.occupied[tx][ty]
            self.occupied[tx][ty] = 0
            self.tail = (ntx, nty)
        
        else:
            tx, ty = self.tail
            if nx == tx and ny == ty:
                ntx, nty = self.occupied[tx][ty]
                self.occupied[hx][hy] = (nx, ny)
                self.occupied[tx][ty] = 0
                self.head = (nx, ny)
                self.tail = (ntx, nty)
            else:
                return -1
        
            
        return self.length
        


# Your SnakeGame object will be instantiated and called as such:
# obj = SnakeGame(width, height, food)
# param_1 = obj.move(direction)
```
