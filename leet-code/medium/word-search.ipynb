# First solution

Problem here is that I don't stop looking in other directions once I've found a match in an initial diretion. This cause it to be too slow for leetcode submission

```python
class Solution:
    def exist(self, board: List[List[str]], word: str) -> bool:
        
        m = len(board)
        n = len(board[0])
        last_dir = -1
        base_path = []
        for i in range(m):
            for j in range(n):
                used = [[0 for u in range(n)] for p in range(m)]
                res = self.search_helper(i,j,board, word,used)
                if res:
                    return res
                    
        
        return False
        
    
    def search_helper(self,i,j,board, word, used):
        #print(last_dir)
        if len(word) == 1 and board[i][j] == word:
            #print('h')
            return True
        
        
        front = word[0]
        rest = word[1:]
        
        if board[i][j] != front:
            return False
        #print(type(used)) 
        used[i][j] = 1
        #search left
        si = i
        sj = j-1
        if self.is_valid(si,sj,board) and not used[si][sj] == 1:
            #print('h1')
            res_l = self.search_helper(si,sj,board,rest,used)
        else:
            res_l = False
                
        #search right
        si = i
        sj = j+1
        if self.is_valid(si,sj,board) and not used[si][sj] == 1:
            #print('h2')
            res_r = self.search_helper(si,sj,board,rest,used)
        else:
            res_r = False
        
        #search up
        si = i-1
        sj = j
        if self.is_valid(si,sj,board) and not used[si][sj] == 1:
            #print('h3')
            res_u = self.search_helper(si,sj,board,rest,used)
        else:
            res_u = False
            
        #search down
        si = i+1
        sj = j
        if self.is_valid(si,sj,board) and not used[si][sj] == 1:
            #print('h4')
            res_d = self.search_helper(si,sj,board,rest,used)
        else:
            res_d = False
        
        s_or = res_l or res_r or res_u or res_d
        if not s_or:
            used[i][j] = 0
            
        return s_or
    
    
    
    def is_valid(self,i,j,board):
        m = len(board)
        n = len(board[0])
        
        if i < 0 or j < 0 or i >= m or j >= n:
            return False
        return True
```

# Better solution

```python
class Solution:
    def exist(self, board: List[List[str]], word: str) -> bool:
        
        m = len(board)
        n = len(board[0])
        last_dir = -1
        base_path = []
        for i in range(m):
            for j in range(n):
                if board[i][j] == word[0]:
                   
                    res = self.search_helper(i,j,board,m,n, word)
                    if res:
                        return res

        
        return False
        
    
    def search_helper(self,i,j,board,m,n,word):
        #print(last_dir)
        if len(word) == 1 and board[i][j] == word[0]:
            #print('h')
            return True
        
        
        front = word[0]
        rest = word[1:]
        
        if board[i][j] != front:
            return False
        #print(type(used)) 
        board[i][j] = '#'
        #search left
        si = i
        sj = j-1
        if self.is_valid(si,sj,m,n,board):
            #print('h1')
            res_l = self.search_helper(si,sj,board,m,n,rest)
            if res_l:
                board[i][j] = word[0]
                return True
        else:
            res_l = False
                
        #search right
        si = i
        sj = j+1
        if self.is_valid(si,sj,m,n,board):
            #print('h2')
            res_r = self.search_helper(si,sj,board,m,n,rest)
            if res_r:
                board[i][j] = word[0]
                return True
        else:
            res_r = False
        
        #search up
        si = i-1
        sj = j
        if self.is_valid(si,sj,m,n,board):
            #print('h3')
            res_u = self.search_helper(si,sj,board,m,n,rest)
            if res_u:
                board[i][j] = word[0]
                return True
        else:
            res_u = False
            
        #search down
        si = i+1
        sj = j
        if self.is_valid(si,sj,m,n,board):
            #print('h4')
            res_d = self.search_helper(si,sj,board,m,n,rest)
            if res_d:
                board[i][j] = word[0]
                return True
        else:
            res_d = False
        
        s_or = res_l or res_r or res_u or res_d
        
        board[i][j] = word[0]
            
        return s_or
    
    
    
    def is_valid(self,i,j,m,n,board):
        
        if i < 0 or j < 0 or i >= m or j >= n or board[i][j] == '#':
            return False
        return True
```
