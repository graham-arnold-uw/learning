```python
 class Solution:
    def findDiagonalOrder(self, mat: List[List[int]]) -> List[int]:
        m = len(mat)
        n = len(mat[0])
        num_to_proc = n*m
        num_processed = 0
         
        level = 0
        direction = 0 #0 = bottom to top | 1 = top to bottom
        
        dir_moves = [(-1,1), (1,-1)] #b2t, t2b
        reset_move = [(1,0), (0,1)]
        out = []
        
        rev = 1
        cx, cy = 0, 0
        
        #add all diagnonals from top row
        for col in range(n):
            cx, cy = 0, col
            ps = []
            while cx >= 0 and cx < m and cy >= 0 and cy < n:
                ps.append(mat[cx][cy])
                cx = cx + 1
                cy = cy - 1
            
            if rev == 1:
                ps = ps[::-1]
            
            out.extend(ps)
            rev = not rev
            
        
        #add all diagonals from last columns
        for row in range(m):
            cx, cy = 1 + row, n - 1
            ps = []
            while cx >= 0 and cx < m and cy >= 0 and cy < n:
                ps.append(mat[cx][cy])
                cx = cx + 1
                cy = cy - 1
            
            if rev == 1:
                ps = ps[::-1]
            
            out.extend(ps)
            rev = not rev
            
        return out
```
