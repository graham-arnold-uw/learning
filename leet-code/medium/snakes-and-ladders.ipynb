# Too slow recursive solution

```python
class Solution:
    def snakesAndLadders(self, board: List[List[int]]) -> int:
        n = len(board)
        
        path = set()
        memo = {}
        min_moves = self.sl_helper(board, 1, n, path, memo)
        return min_moves if min_moves < float('inf') else -1
    
    
    def sl_helper(self, board, pos, n, path, memo):
        
        if pos == n * n:
            return 0
        
        elif pos in memo:
            return memo[pos]
        
        elif pos in path:
            return float('inf')
        
        path.add(pos)
        options = range(pos + 1, min(pos + n, n*n) + 1)
        min_path = float('inf')
        for opt in options:
            i, j = self.convert_from_boust(opt, n)
            if board[i][j] != -1:
                min_next = self.sl_helper(board, board[i][j], n, path, memo)
            else:
                min_next = self.sl_helper(board, pos + 1, n, path, memo)
            
            min_path = min(min_path, min_next)
        
        path.remove(pos)
        return min_path + 1 if min_path < float('inf') else float('inf')
        
        
    def convert_to_boust(self, i, j, n):
            rows_from_bottom = n - 1 - i

            boust = 0
            boust += (rows_from_bottom * n)

            if rows_from_bottom % 2 == 0:
                boust += j

            else:
                boust += n - 1 - j

            boust += 1
            return boust
    
    def convert_from_boust(self, b, n):
        b -= 1

        row_count = b // n
        excess = b % n

        row = n - 1 - row_count
        if row_count % 2 != 0:
            col = n - 1 - excess
        else:
            col = excess

        return (row, col)
```

# BFS solution

```python
from collections import deque

class Solution:
    def snakesAndLadders(self, board: List[List[int]]) -> int:
        n = len(board)
        q = deque()
        
        min_path = float('inf')
        
        visited = set()
        visited.add(1)
        q.append((1, 0))
        
        while len(q) > 0:
            #print(q)
            curr_pos, curr_moves = q.popleft()
            
            #print(f"curr_ pos: {curr_pos} curr_moves: {curr_moves}")
            
            if curr_pos == n * n:
                return curr_moves
            
            
            for k in range(1, 7):
                next_pos = curr_pos + k
                if next_pos > n*n:
                    break
                
                ni, nj = self.convert_from_boust(next_pos, n)
                if board[ni][nj] != -1:
                    next_pos = board[ni][nj]
                
                if next_pos not in visited:
                    q.append((next_pos, curr_moves + 1))
                    visited.add(next_pos)
                
        
        return -1
        
        
    def convert_to_boust(self, i, j, n):
            rows_from_bottom = n - 1 - i

            boust = 0
            boust += (rows_from_bottom * n)

            if rows_from_bottom % 2 == 0:
                boust += j

            else:
                boust += n - 1 - j

            boust += 1
            return boust
    
    def convert_from_boust(self, b, n):
        b -= 1

        row_count = b // n
        excess = b % n

        row = n - 1 - row_count
        if row_count % 2 != 0:
            col = n - 1 - excess
        else:
            col = excess

        return (row, col)
```
