```python
class Solution:
    def nextClosestTime(self, time: str) -> str:
        
        def backtracking(path, avail, i, n):
            if i == n:
                #curr_mins = get_minutes(path)
                curr_delta = get_time_diff(original, path)
                #print(f"currdelta = {curr_delta}")
                #print(f"time = {path}")
                #print()
                if curr_delta < closest_time[0] and curr_delta > 0:
                    closest_time[0] = curr_delta
                    closest_string[0] = list_to_string(path)
                    #print(f"currdelta = {curr_delta}")
                    #print(f"time = {closest_string[0]}")
                    #print()
                return
            
            for av in avail:
                if is_valid(av, i, path):
                    path.append(av)
                    backtracking(path, avail, i + 1, n)
                    path.pop()
            return
        
        def get_minutes(t1_list):
            min_since1 = 0
            num_hours1 = t1_list[0] * 10 +t1_list[1]
            num_mins1 = t1_list[2] * 10 +t1_list[3]
            
            min_since1 = 60 * num_hours1 + num_mins1
            return min_since1
        
        #t1 always the original time
        def get_time_diff(t1_list, t2_list):
            min1 = get_minutes(t1_list)
            min2 = get_minutes(t2_list)
            
            
            if min2 >= min1:
                diff = min2 - min1
            else:
                left = 24 * 60 - min1
                diff = min2 + left
            return diff
        
        def list_to_string(t_list):
            h1, h2 = str(t_list[0]), str(t_list[1])
            m1, m2 = str(t_list[2]), str(t_list[3])
            return h1+h2+':'+m1+m2
        
        def is_valid(num, i, path):
            if i == 0:
                if not (0 <= num <= 2):
                    return False
            elif i == 1:
                if path[-1] == 1 or path[-1] == 0:
                    if not (0 <= num <= 9):
                        return False
                elif path[-1] == 2:
                    if not (0 <= num < 4):
                        return False
            elif i == 2:
                if not (0 <= num < 6):
                    return False
            elif i == 3:
                if not (0 <= num <= 9):
                    return False
            return True
        
        hours, minutes = time.split(':')
        h1,h2 = hours
        m1,m2 = minutes
        
        original = [int(h1), int(h2), int(m1), int(m2)]
        original_minutes = get_minutes(original)
        
        avail = {int(h1), int(h2), int(m1), int(m2)}
        
        closest_time = [float('inf')]
        closest_string = [""]
        
        if int(h1) == int(h2) == int(m1) == int(m2):
            return time
        path = []
        backtracking(path, avail, 0, 4)
        return closest_string[0]
        
        
        
        
        
```
