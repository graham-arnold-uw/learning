```python
class Node:
    def __init__(self, val = 0, neighbors = None):
        self.val = val
        self.neighbors = neighbors if neighbors is not None else []

class Solution:
        
    
    def canFinish(self, numCourses: int, prerequisites: List[List[int]]) -> bool:
        
        node_list = self.create_graphs(numCourses, prerequisites)
        #self.print_graph(node_list)
        
        visited = [False] * numCourses
        rec_stack = [False] * numCourses
        
        for node_i in node_list:
            if visited[node_i.val] == False:
                if self.check_cycle_util(node_i, visited, rec_stack):
                    return False
        
        return True
        
    def check_cycle_util(self, node, visited, rec_stack):
        
        visited[node.val] = True
        rec_stack[node.val] = True
        
        for n in node.neighbors:
            if not visited[n.val]:
                res = self.check_cycle_util(n,visited,rec_stack)
                if res:
                    return True
            elif rec_stack[n.val]:
                return True
        
        rec_stack[node.val] = False
        return False
    
    
    
    def create_graphs(self, numCourses, prereqs):
        
        node_list = []
        for i in range(numCourses):
            node_i = Node(i, [])
            node_list.append(node_i)
        
        for p in prereqs:
            node_list[p[0]].neighbors.append(node_list[p[1]])
        
        return node_list
    
    def print_graph(self, node_list):
        
        for node in node_list:
            print('processing node ' + str(node.val))
            
            for n in node.neighbors:
                print(str(n.val) + ' ', end='')
            print()
    

        
```
