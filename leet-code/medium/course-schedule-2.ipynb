```python
from collections import defaultdict
from collections import deque

class Solution:
    
    def findOrder(self, numCourses: int, prerequisites: List[List[int]]) -> List[int]:
        
        res = deque()
        
        def top_sort(dag, num_courses):
            visited = [0 for i in range(num_courses)]
            
            for course in range(num_courses):
                if visited[course] != 1:
                    dfs(course, dag, visited)
                    
        def check_cycle(graph, num_courses):
            visited = [False for i in range(num_courses)]
            rec_stack = [False for i in range(num_courses)]
            
            for course in range(num_courses):
                if visited[course] == False:
                    if cycle_util(graph, course, visited, rec_stack):
                        return True
            return False
        
        def cycle_util(graph, node, visited, rec_stack):
            visited[node] = True
            rec_stack[node] = True
            
            for neighbor in graph[node]:
                if not visited[neighbor]:
                    if cycle_util(graph, neighbor, visited, rec_stack):
                        return True
                elif rec_stack[neighbor]:
                    return True
            
            rec_stack[node] = False
            return False
        
        def dfs(start, graph, visited):
            
            visited[start] = 1
            children = graph[start]
            for child in children:
                if visited[child] != 1:
                    dfs(child, graph, visited)
            
            res.appendleft(start)
            return
        
        def build_graph(prereqs, num_courses):
            g = defaultdict(list)
            
            for i in range(num_courses):
                for j in range(len(prereqs)):
                    if prereqs[j][1] == i:
                        g[i].append(prereqs[j][0])

            return g
        
        
        dag = build_graph(prerequisites, numCourses)
        
        is_cycle = check_cycle(dag, numCourses)
        
        if is_cycle:
            return []
        
        top_sort(dag, numCourses)
        
        res = list(res)
        
        return res
        
```
