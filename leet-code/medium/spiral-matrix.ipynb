# Final solution


```python
class Solution:
    def spiralOrder(self, matrix: List[List[int]]) -> List[int]:
        
        m = len(matrix)
        n = len(matrix[0])
        
        
        tot = m*n
        seen = 0
        
        i = j = 0
        
        res = []
        
        ul = [i,j]
        lr = [i+m-1, j+n-1]
        strt = [i,j]
        
        itr = 0
        
        while seen < tot:
            
            d = itr % 4
            
            w = lr[1] - ul[1] + 1
            h = lr[0] - ul[0] + 1
            
            #print(w)
            #print(h)
            #print(d)
            
            if d == 0 or d == 2:
                mv_ct = w
            else:
                mv_ct = h
                #print(mv_ct)
                
            self.move_dir(strt,matrix,d,mv_ct,res)
            
            seen += mv_ct
            
            if d == 0:
                strt[0] += 1
                strt[1] += (mv_ct - 1)
                ul[0] += 1
            elif d == 1:
                strt[1] -= 1
                strt[0] += (mv_ct - 1)
                lr[1] -= 1
            elif d == 2:
                strt[0] -= 1
                strt[1] -= (mv_ct -1)
                lr[0] -= 1
            else:
                strt[1] += 1
                strt[0] -= (mv_ct - 1)
                ul[1] += 1
                
            itr += 1
    
        return res
    
    
    #right - > upper right of next square is start
    #down -> lower right of next sq
    #left -> lower left
    #up -> upper left
    
    
    def move_dir(self,start ,mat,d, ct,out):
        i = start[0]
        j = start[1]
        #print(ct)
        for k in range(ct):
            #print(i)
            #print(j)
            out.append(mat[i][j])
            if d == 0:
                j += 1
            elif d == 1:
                i += 1
            elif d == 2:
                j -= 1
            else:
                i -= 1
            #print()
        
```
