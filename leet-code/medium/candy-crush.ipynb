# Read head write head approach for moving zeroes to top

```python
def move_zeroes_to_top(board):
    m = len(board)
    n = len(board[0])
    
    for col in range(n):
        read = write = m - 1
        
        while read >= 0:
            if board[read][col] != 0:
                board[write][col] = board[read][col]
                write -= 1
            read -= 1
        
        while write >= 0:
            board[write][col] = 0
            write -= 1
        
    print(board)
```

```python
def convert_to_boust(i, j, n):
        rows_from_bottom = n - 1 - i
        
        boust = 0
        boust += (rows_from_bottom * n)
        
        if rows_from_bottom % 2 == 0:
            boust += j
            
        else:
            boust += n - 1 - j
        
        boust += 1
        return boust
    
def convert_from_boust(b, n):
    b -= 1

    row_count = b // n
    excess = b % n

    row = n - 1 - row_count
    if row_count % 2 != 0:
        col = n - 1 - excess
    else:
        col = excess

    return (row, col)

    
```
