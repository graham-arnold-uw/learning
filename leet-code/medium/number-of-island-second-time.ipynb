```python
from collections import deque

class Solution:
    def numIslands(self, grid: List[List[str]]) -> int:
        m = len(grid)
        n = len(grid[0])
        visited = self.create_visited(m,n,grid)
        #discovered = self.create_visited(m,n,grid)
        isl_count = 0
    
        for i in range(m):
            for j in range(n):
                if grid[i][j] == "1" and visited[i][j] == 0:
                    self.search_island(i,j,grid,visited)
                    isl_count += 1
        
        return isl_count
    
    def search_island(self,x,y,grid,visited):
        pt = (x,y)
        visited[x][y] = 1
        
        q = deque()
        q.append(pt)
        
        while len(q) > 0:
            cx,cy = q.popleft()
            unv = self.find_neighbors(cx,cy,grid,visited)
            
            for n in unv:
                cx,cy = n
                visited[cx][cy] = 1
                q.append(n)
    
    def find_neighbors(self,x,y,grid,visited):
        m = len(grid)
        n = len(grid[0])        
        unv = []
        
        directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]
        
        for dx, dy in directions:
            cx, cy = x + dx, y + dy
            if cx >= 0 and cx < m and cy >= 0 and cy < n and grid[cx][cy] == '1' and visited[cx][cy] == 0:
                unv.append((cx, cy))
        
        return unv
            
        
    
    def create_visited(self,m,n,grid):
        visited = [[-1 for j in range(n)] for i in range(m)]
            
        for i in range(m):
            for j in range(n):
                if grid[i][j] == '1':
                    visited[i][j] = 0
        return visited
           
            
        
        

        
        
        
        
```
