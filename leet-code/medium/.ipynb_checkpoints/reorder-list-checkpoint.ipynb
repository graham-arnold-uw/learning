```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def reorderList(self, head: ListNode) -> None:
        """
        Do not return anything, modify head in-place instead.
        """
        
        list_len = self.get_list_len(head)
        if list_len % 2 == 0:
            mid = list_len / 2
        else:
            mid = (list_len // 2) + 1
        
        f_half, l_half = self.split_list(head,mid)
        
        #print(f_half)
        #print(l_half)
        l_rev = self.reverse_list(l_half)
        
        merged = self.merge_lists(f_half, l_rev)
        
        return merged
    
    
    def merge_lists(self, l1, l2):
        
        if l1 == None:
            return l2
        if l2 == None:
            return l1
        
        new_head = l1
        c1 = l1
        c2 = l2
        
        while c1 != None and c2 != None:
            t1 = c1.next
            t2 = c2.next
            '''
            if c2 != None:
                t2 = c2.next
            else:
                t2 = None'''
            
            c1.next = c2
            c2.next = t1
            
            c1 = t1
            c2 = t2
        
        return new_head
    
    def get_list_len(self,head):
        count = 0
        
        while head != None:
            count += 1
            head = head.next
        
        return count
    
    def split_list(self,head,mid):
        
        ct = 0
        f_half = head
        l_half = head
        
        while ct < mid:
            if ct == mid - 1:
                t = l_half.next
                l_half.next = None
                l_half = t
            else:
                l_half = l_half.next
            ct += 1
        
        return f_half, l_half
    
    
    def reverse_list(self, head):
        
        if head == None:
            return None
        
        prev = None
        curr = head
        
        while curr != None:
            temp = curr.next
            curr.next = prev
            prev = curr
            curr = temp
        
        return prev
            
```
