# Solution but slow

```python
class Solution:
    def candyCrush(self, board: List[List[int]]) -> List[List[int]]:
        
        m = len(board)
        n = len(board[0])
        
        crush_found = self.get_crush_groups(board, m, n)
        
        #if not crush_found:
            #return board
        
        while crush_found:
            
            self.perform_shift(board, m, n)
            
            crush_found = self.get_crush_groups(board, m, n)
            #set crush flagged elements to zero
            
       
        return board
            
    
    def perform_shift(self, board, m, n):
        
        for col in range(n):    
            for row in range(m):
                
                if board[row][col] == 0:
                    self.bubble_up(board, row, col, m, n)
    
    def bubble_up(self, board, row, col, m, n):
        
        while row > 0 and board[row][col] < board[row - 1][col]:
            board[row][col], board[row - 1][col] = board[row - 1][col], board[row][col]
            row -= 1
        
    def get_crush_groups(self, board, m, n):
        
        crush_candidates = set()
        h_visited = [[0 for j in range(n)] for i in range(m)]
        v_visited = [[0 for j in range(n)] for i in range(m)]
        #crush_mat = [[0 for j in range(n)] for i in range(m)]
        
        
        for i in range(m):
            for j in range(n):
                
                curr_el = board[i][j]
                
                if curr_el != 0:
                    horz_matches = set()
                    vert_matches = set()

                    if h_visited[i][j] != 1:
                        for k in range(j, n):
                            if board[i][k] == curr_el:
                                h_visited[i][k] = 1
                                horz_matches.add((i,k))
                            else:
                                break

                    if v_visited[i][j] != 1:
                        for k in range(i, m):
                            if board[k][j] == curr_el:
                                v_visited[k][j] = 1
                                vert_matches.add((k, j))
                            else:
                                break

                    if len(horz_matches) >= 3:
                        crush_candidates.update(horz_matches)

                    if len(vert_matches) >= 3:
                        crush_candidates.update(vert_matches)

                    #find crush candidates for each position
                    #if len horz crush candidates > 3:
                        #mark in crush matrix for removal


        if len(crush_candidates) < 3:
            return 0
        else:
            for ci,cj in crush_candidates:
                board[ci][cj] = 0
            return 1
        
        
```

```python
class Solution:
    def candyCrush(self, board: List[List[int]]) -> List[List[int]]:
        
        m = len(board)
        n = len(board[0])
        
        crush_found = self.get_crush_groups(board, m, n)
        
        #if not crush_found:
            #return board
        
        while crush_found:
            
            self.perform_shift(board, m, n)
            
            crush_found = self.get_crush_groups(board, m, n)
            #set crush flagged elements to zero
            
       
        return board
            
    
    def perform_shift(self, board, m, n):
        
        for col in range(n):    
            for row in range(m):
                
                if board[row][col] == 0:
                    self.bubble_up(board, row, col, m, n)
    
    def bubble_up(self, board, row, col, m, n):
        
        while row > 0 and board[row][col] < board[row - 1][col]:
            board[row][col], board[row - 1][col] = board[row - 1][col], board[row][col]
            row -= 1
      
    def delete_crush_groups(self, board, m, n):
        deleted_mat = [[0 for j in range(n)] for i in range(m)]
        
        #get horz crush groups
        for i in range(m):
            ws = 0
            we = 0
            
            while we < n:
                if board[i][we] != 0:
                    
                    if board[i][we] != board[i][ws]:
                        w_len = we - ws

                        if w_len >= 3:
                            for g in range(ws, ws + w_len):
                                deleted_mat[i][g] = 1

                        ws = we
                    elif board[i][ws] == board[i][we]:
                        we += 1
                else:
                    we += 1
                    ws = we
                    
            w_len = we - ws
            if w_len >= 3:
                for g in range(ws, ws + w_len):
                            deleted_mat[i][g] = 1
        
        #get vert crush groups
        for i in range(n):
            ws = 0
            we = 0
            
            while we < m:
                if board[we][i] != 0:
                    if board[we][i] != board[ws][i]:
                        w_len = we - ws

                        if w_len >= 3:
                            for g in range(ws, ws + w_len):
                                deleted_mat[g][i] = 1

                        ws = we
                    elif board[ws][i] == board[we][i]:
                        we += 1
                else:
                    we += 1
                    ws = we
                 
                                                 
            w_len = we - ws
            if w_len >= 3:
                for g in range(ws, ws + w_len):
                            deleted_mat[g][i] = 1
                        
        deleted = False
        
        for i in range(m):
            for j in range(n):
                if deleted_mat[i][j] == 1:
                    board[i][j] = 0
                    deleted = True
                    
        return deleted
        
    def get_crush_groups(self, board, m, n):
        
        crush_candidates = set()
        h_visited = [[0 for j in range(n)] for i in range(m)]
        v_visited = [[0 for j in range(n)] for i in range(m)]
        #crush_mat = [[0 for j in range(n)] for i in range(m)]
        
        
        for i in range(m):
            for j in range(n):
                
                curr_el = board[i][j]
                
                if curr_el != 0:
                    horz_matches = set()
                    vert_matches = set()

                    if h_visited[i][j] != 1:
                        for k in range(j, n):
                            if board[i][k] == curr_el:
                                h_visited[i][k] = 1
                                horz_matches.add((i,k))
                            else:
                                break

                    if v_visited[i][j] != 1:
                        for k in range(i, m):
                            if board[k][j] == curr_el:
                                v_visited[k][j] = 1
                                vert_matches.add((k, j))
                            else:
                                break

                    if len(horz_matches) >= 3:
                        crush_candidates.update(horz_matches)

                    if len(vert_matches) >= 3:
                        crush_candidates.update(vert_matches)

                    #find crush candidates for each position
                    #if len horz crush candidates > 3:
                        #mark in crush matrix for removal


        if len(crush_candidates) < 3:
            return 0
        else:
            for ci,cj in crush_candidates:
                board[ci][cj] = 0
            return 1
        
        
```
