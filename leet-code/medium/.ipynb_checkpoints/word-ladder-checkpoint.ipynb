```python
from collections import deque

class Node:
    
    def __init__(self, val):
        self.val = val
        self.children = []

class Solution:
    def ladderLength(self, beginWord: str, endWord: str, wordList: List[str]) -> int:
        
        root = Node(beginWord)
        
        self.build_tree(root, wordList)
        
        q = deque()
        q.append([1, root])
        
        while len(q) > 0:
            pathlen, curr_node = q.popleft()
            if curr_node.val == endWord:
                return pathlen
            
            for c in curr_node.children:
                q.append([pathlen + 1, c])
        
        return 0
        
    
    def build_tree(self, root, wordlist):
        q = deque()
        n = len(wordlist)
        
        q.append(root)
        parents = {root.val}
        
        while len(q) > 0:
            curr_node = q.popleft()
            for i in range(n):
                if wordlist[i] not in parents and self.one_diff(wordlist[i], curr_node.val):
                    #make a child with this word[i] val
                    new_child = Node(wordlist[i])
                    curr_node.children.append(new_child)
                    q.append(new_child)
                    parents.add(wordlist[i])
    
     
    def one_diff(self, s1, s2):
        diff_used = False
        n = len(s1)
        i = 0
        
        while i < n:
            if s1[i] != s2[i]:
                if diff_used:
                    return False
                else:
                    diff_used = True
            i += 1
        return True
            
            
            
            
            
            
            
            
            
            
            
            
            
            
            
            
            
            
            
            
```
