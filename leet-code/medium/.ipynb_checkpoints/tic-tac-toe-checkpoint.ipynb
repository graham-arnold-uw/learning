# Slow O(N^2) solution using searching technique

```python
class TicTacToe:

    def __init__(self, n: int):
        self.n = n
        self.board = [[0 for i in range(self.n)] for j in range(self.n)]
        

    def move(self, row: int, col: int, player: int) -> int:
        self.board[row][col] = player
        
        res = self.check_board()
        
        return res if res else 0
    
    
    def check_board(self):
        #check rows down left side
        for row in range(self.n):
            first = self.board[row][0]
            
            if first == 0:
                continue
                
            col = 0
            while col < self.n and self.board[row][col] == first:
                col += 1
            
            if col == self.n:
                #print('row true')
                return first
        
        
        #check columns down top
        for col in range(self.n):
            first = self.board[0][col]
            
            if first == 0:
                continue
                
            row = 0
            while row < self.n and self.board[row][col] == first:
                row += 1
            
            if row == self.n:
                #print('col true')
                return first
        
        #check top diag
        first = self.board[0][0]
        ct = 0
        
        if first != 0:
            while ct < self.n and self.board[ct][ct] == first:
                ct += 1
        
        if ct == self.n:
                #print('td true')
                return first
            
        #check bottom diag
        first = self.board[self.n - 1][0]
        ct = 0
        
        if first != 0:
            while ct < self.n and self.board[self.n - ct - 1][ct] == first:
                ct += 1
        
        if ct == self.n:
                #print('bd true')
                return first
        
        return 0

# Your TicTacToe object will be instantiated and called as such:
# obj = TicTacToe(n)
# param_1 = obj.move(row,col,player)
```

# Optimized Solution using Hash Tables


```python
from collections import defaultdict
class TicTacToe:

    def __init__(self, n: int):
        self.n = n
        #self.board = [[0 for i in range(self.n)] for j in range(self.n)]
        
        #self.row_cts = defaultdict(int)
        #self.col_cts = defaultdict(int)
        #self.diag_cts = defaultdict(int)
        
        self.player_cts = {1: [defaultdict(int), defaultdict(int), defaultdict(int)], 2: [defaultdict(int), defaultdict(int), defaultdict(int)] }
        

    def move(self, row: int, col: int, player: int) -> int:
        #self.board[row][col] = player
        
        player_ct_list = self.player_cts[player]
        
        row_dict, col_dict, diag_dict = player_ct_list
        
        row_dict[row] += 1
        if row_dict[row] == self.n:
            return player
        
        col_dict[col] += 1
        if col_dict[col] == self.n:
            return player
        
        if row == col:
            diag_dict[0] += 1
            if diag_dict[0] == self.n:
                return player
        
        if row + col == self.n - 1:
            diag_dict[1] += 1
            if diag_dict[1] == self.n:
                return player
        
        return 0
    
    

# Your TicTacToe object will be instantiated and called as such:
# obj = TicTacToe(n)
# param_1 = obj.move(row,col,player)
```
