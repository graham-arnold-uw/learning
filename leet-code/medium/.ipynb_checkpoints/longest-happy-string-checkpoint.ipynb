```python
import heapq

class Solution:
    def longestDiverseString(self, a: int, b: int, c: int) -> str:
        char_freqs = {}
        occurences  = []
        if a > 0:
            occurences.append([-a, 'a'])
        if b > 0:
            occurences.append([-b, 'b'])
        if c > 0:
            occurences.append([-c, 'c'])
        
        sorted_occurences = sorted(occurences, key = lambda x : x[0])
        #print(sorted_occurences)
        #total = a + b + c
        heapq.heapify(sorted_occurences)
        #print(sorted_occurences)
        res = []
        
        prev_largest = heapq.heappop(sorted_occurences)
        #print(prev_largest)
        
        for i in range(min(-prev_largest[0], 2)):
            res.append(prev_largest[1])
            prev_largest[0] += 1
        
        while len(sorted_occurences) > 0:
            #print(sorted_occurences)
            #print()
            next_largest = heapq.heappop(sorted_occurences)
            next_left, next_char = -next_largest[0], next_largest[1]
            for i in range(min(next_left, 2)):
                res.append(next_char)
                next_largest[0] += 1
                
            if prev_largest[0] < 0:
                heapq.heappush(sorted_occurences, prev_largest)
                
            prev_largest = next_largest
        
        #print(prev_largest)
        while prev_largest[0] < 0:
            next_index = self.find_index(res, prev_largest[1])
            if next_index != -1:
                for i in range(min(-prev_largest[0],2)):
                    res.insert(next_index, prev_largest[1])
                    prev_largest[0] += 1
            else:
                break
            
        out  = "".join(res)
        return out
    
    def find_index(self, char_list, tar):
        n = len(char_list)
        
        i = 0
        
        while i < n - 1:
            if char_list[i] != tar and char_list[i + 1] != tar:
                return i + 1
            i += 1
        
        return - 1
            
                
                
        
        
        
```
