# Time limit exceeded - backtracking solution

```python
class Solution:
    def pacificAtlantic(self, matrix: List[List[int]]) -> List[List[int]]:
        mlen = len(matrix)
        
        if mlen == 0:
            return []
        nlen = len(matrix[0])
        
        res = []
        
        for i in range(mlen):
            for j in range(nlen):
                visited_pac = [[False for k in range(nlen)] for q in range(mlen)]
                visited_atl = [[False for k in range(nlen)] for q in range(mlen)]
                try_pac = self.check_path(i,j,matrix,0, visited_pac)
                try_atl = self.check_path(i,j,matrix,1, visited_atl)
                
                if try_pac and try_atl:
                    res.append([i,j])
            
        
        return res
    
    
    
    
    def check_path(self, i, j, mat, choice, visited):
        
        mlen = len(mat)
        nlen = len(mat[0])
        
        if choice == 0:
            if i == 0 or j == 0:
                return True
        elif choice == 1:
            if i == mlen - 1 or j == nlen - 1:
                return True
        
        visited[i][j] = True
        
        
        #try left
        nxti = i
        nxtj = j-1
        
        if self.is_valid_move(i,j,nxti,nxtj,mat,visited):
            try_l = self.check_path(nxti,nxtj,mat,choice,visited)
            if try_l:
                return True
            
        #try right
        nxti = i
        nxtj = j+1
        
        if self.is_valid_move(i,j,nxti,nxtj,mat,visited):
            try_r = self.check_path(nxti,nxtj,mat,choice,visited)
            if try_r:
                return True
            
        #try up
        nxti = i-1
        nxtj = j
        
        if self.is_valid_move(i,j,nxti,nxtj,mat,visited):
            try_u = self.check_path(nxti,nxtj,mat,choice,visited)
            if try_u:
                return True
            
         #try down
        nxti = i+1
        nxtj = j
        
        if self.is_valid_move(i,j,nxti,nxtj,mat,visited):
            try_d = self.check_path(nxti,nxtj,mat,choice,visited)
            if try_d:
                return True
        
        visited[i][j] = False
        return False
            
            
            
            
    def is_valid_move(self,i,j,nxti,nxtj,mat,visited):
        mlen = len(mat)
        nlen = len(mat[0])
        
        if nxti >= 0 and nxti < mlen and nxtj >= 0 and nxtj < nlen and visited[nxti][nxtj] != True and mat[nxti][nxtj] <= mat[i][j]:
            return True
        return False
```

```python
from collections import deque
class Solution:
    def pacificAtlantic(self, matrix: List[List[int]]) -> List[List[int]]:
        mlen = len(matrix)
        
        if mlen == 0:
            return []
        nlen = len(matrix[0])
        
        res = []
        sol_pac = [[-1 for k in range(nlen)] for q in range(mlen)]
        sol_atl = [[-1 for k in range(nlen)] for q in range(mlen)]
        
        for i in range(mlen):
            for j in range(nlen):
                curr = matrix[i][j]
                if sol_pac[i][j] == -1:
                    path_pac = self.bfs_ocean(matrix,i,j, 0)
                    #print(path_pac)
                    if path_pac == -1:
                        sol_pac[i][j] = False
                    else:
                        for coord in path_pac:
                            ci,cj = coord
                            sol_pac[ci][cj] = True

                    #perform BFS from coord,
                    #return path to dest ocean if exists
                    # else return -1 
                    #add all elements in path to sol
                if sol_atl[i][j] == -1:
                    #print('atl')
                    path_atl = self.bfs_ocean(matrix,i,j,1)
                    #if i == 0 and j == 4:
                        #print(path_atl)

                    #print(path_atl)
                    if path_atl == -1:
                        sol_atl[i][j] = False
                    else:
                        for coord in path_atl:
                            ci,cj = coord
                            if ci == 0 and cj == 3:
                                print(str(i)+' '+str(j))
                                print(path_atl)
                            sol_atl[ci][cj] = True
                            
                    
                     #perform BFS from coord,
                    #return path to dest ocean if exists
                    # else return -1 
                    #add all elements in path to sol

        
        #print(sol_pac)
        #print(sol_atl)
        for i in range(mlen):
            for j in range(nlen):
                if sol_pac[i][j] == True and sol_atl[i][j] == True:
                    res.append([i,j])
        return res
    
    
    
    
    def bfs_ocean(self,mat,i,j, dest):
        
        mlen = len(mat)
        nlen = len(mat[0])
        q = deque()
        visited = [[False for k in range(nlen)] for q in range(mlen)]
        
        init_path = [(i,j)]
        
    
        q.append(init_path)
        
        while len(q) > 0:
            #print(q)
            curr_path = q.popleft()
            #print(curr_path)
            curr_i, curr_j = curr_path[-1]
            visited[curr_i][curr_j] = True
            if self.is_ocean(mat,dest,curr_i,curr_j):
                return curr_path
            
            adj = self.find_flow(mat,curr_i,curr_j,visited)
            
            for a in adj:
                temp = curr_path.copy()
                temp.append(a)
                q.append(temp)
                #if i == 1 and j == 3:
                    #print(q)
        
        return -1
        
        
            
            
    
            
    def is_ocean(self,mat,dest,i,j):
        mlen = len(mat)
        nlen = len(mat[0])
        
        if dest == 0:
            if i == 0 or j == 0:
                return True
            else:
                return False
        else:
            if i == mlen-1 or j == nlen-1:
                return True
            else:
                return False
            
            
    def find_flow(self,mat,i,j,visited):
        adj = []
        
        #try up
        nxti = i-1
        nxtj = j
        if self.is_valid_move(i,j,nxti,nxtj,mat,visited):
            adj.append((nxti,nxtj))
            
        #try down
        nxti = i+1
        nxtj = j
        if self.is_valid_move(i,j,nxti,nxtj,mat,visited):
            adj.append((nxti,nxtj))
            
        #try left
        nxti = i
        nxtj = j-1
        if self.is_valid_move(i,j,nxti,nxtj,mat,visited):
            adj.append((nxti,nxtj))
        
        #try right
        nxti = i
        nxtj = j+1
        if self.is_valid_move(i,j,nxti,nxtj,mat,visited):
            adj.append((nxti,nxtj))
        
        return adj
            
        
        
    def is_valid_move(self,i,j,nxti,nxtj,mat,visited):
        mlen = len(mat)
        nlen = len(mat[0])
        
        if nxti >= 0 and nxti < mlen and nxtj >= 0 and nxtj < nlen and visited[nxti][nxtj] != True and mat[nxti][nxtj] <= mat[i][j]:
            return True
        return False            
```
