```python
class TrieNode:
    
    def __init__(self):
        self.children = [None for i in range(26)]
        self.end_of_word = False
         

class Trie:

    def __init__(self):
        self.root = TrieNode()

    def insert(self, word: str) -> None:
        wl = len(word)
        curr_node = self.root
        for i in range(wl):
            lt = word[i]
            lt_idx = self.get_letter_idx(lt)
            
            if curr_node.children[lt_idx] == None:
                curr_node.children[lt_idx] = TrieNode()
                
            curr_node = curr_node.children[lt_idx]
        
        curr_node.end_of_word = True
        
    def get_letter_idx(self, letter):
        return ord(letter) - ord('a')
    
    def search2(self, word: str) -> bool:
        wl = len(word)
        curr_node = self.root
        for i in range(wl):
            lt = word[i]
            lt_idx = self.get_letter_idx(lt)
            
            if curr_node.children[lt_idx] == None:
                return False
            else:
                curr_node = curr_node.children[lt_idx]
        
        return True if curr_node.end_of_word else False

    def search(self, word: str) -> bool:
        starts_res, end_node = self.startsWithHelper(word)
        if starts_res and end_node.end_of_word:
            return True
    
    def startsWith(self, prefix: str) -> bool:
        res, end_node = self.startsWithHelper(prefix)
        return res
    
    def startsWithHelper(self, prefix: str) -> bool:
        wl = len(prefix)
        curr_node = self.root
        for i in range(wl):
            #print('h')
            lt = prefix[i]
            lt_idx = self.get_letter_idx(lt)
            #print(curr_node.children)
            if curr_node.children[lt_idx] == None:
                return False, None
            else:
                curr_node = curr_node.children[lt_idx]
        
        return True, curr_node


# Your Trie object will be instantiated and called as such:
# obj = Trie()
# obj.insert(word)
# param_2 = obj.search(word)
# param_3 = obj.startsWith(prefix)
```
