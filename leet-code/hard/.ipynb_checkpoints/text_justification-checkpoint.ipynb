https://leetcode.com/problems/text-justification/

```python
class Solution:
    def fullJustify(self, words: List[str], maxWidth: int) -> List[str]:
        
        i = 0
        
        lines = []
        
        while i < len(words): 
            line, i = self.process_line(words, maxWidth, i)
            lines.append(line)
            #print(lines)
        #print(lines)
        last_line = lines.pop()
        #print('ll split')
        #print(last_line.split())
        ll_words = last_line.split()
        #print(ll_words)
        ll_mod = []
        occupied = 0
        for ll in ll_words:
            ll_mod.append(ll)
            ll_mod.append(' ')
            occupied += (len(ll) + 1)
        
        occupied -= 1
        ll_mod = ll_mod[:-1]
        if occupied > maxWidth:
            ll_mod = ll_mod[:maxWidth]
        elif occupied < maxWidth:
            delta = maxWidth - occupied
            for i in range(delta):
                ll_mod.append(' ')
        #ll_mod = ll_mod[:-1]
        #print(ll_mod)
        lines.append("".join(ll_mod))
        
        return lines
    
    
    def process_line(self, words, max_width, i):
        line_words = []
        char_occupied = 0
        char_sum = 0
        
        #handle case when last line is being processes
        while i < len(words) and char_occupied < max_width:
            curr_word = words[i]
            
            if char_occupied + len(curr_word) > max_width:
               # print('hit')
                break
            
            
            line_words.append(curr_word)
            char_sum += len(curr_word)
            char_occupied += len(curr_word)
            if char_occupied < max_width:
                char_occupied += 1
            i += 1
           # print(curr_word)
            #print(char_occupied)
        #print("print line words")
        #print(line_words)
        free_spaces = max_width - char_sum
        num_space_invals = len(line_words) - 1
        if num_space_invals > 0:
            base_inval = free_spaces // num_space_invals
            excess = free_spaces % num_space_invals
        else:
            base_inval = free_spaces
            excess = 0
            num_space_invals = 1
        
        base_space = ''
        for k in range(base_inval):
            base_space = base_space + ' '
            
        spaces = [base_space for k in range(num_space_invals)]
        
        for k in range(excess):
            spaces[k] = spaces[k] + ' '
            
        line_res = []
        for k in range(num_space_invals):
            line_res.append(line_words[k])
            line_res.append(spaces[k])
        
        if len(line_words) > 1:
            line_res.append(line_words[-1])
        
        line_str = "".join(line_res)
        return line_str, i
            
        
        
```
