# Time Limit Exceeded

```python
from collections import defaultdict
class Solution:
    def minAreaRect(self, points: List[List[int]]) -> int:
        y_groups = self.group_by_y(points)
        x_groups = self.group_by_x(points)
        #print(y_groups)
        #print(x_groups)
        #print()
        min_area = float('inf')
        for lvl in y_groups:
            pt_list = y_groups[lvl]
            pt_list.sort()
            n = len(pt_list)
            for i in range(n):
                blx,bly = pt_list[i]
                #print(pt_list[i])
                for j in range(i + 1, n):
                    brx,bry = pt_list[j]
                    #print(pt_list[j])
                    tl_cands = x_groups[blx]
                    tr_cands = x_groups[brx]
                    #print(tl_cands)
                    #print(tr_cands)
                    #print()
                    for tl_cand in tl_cands:
                        tlx, tlyy = tl_cand
                        for tr_cand in tr_cands:
                            trx, tryy = tr_cand
                            if tlyy > bly and tryy > bry and tlyy == tryy:
                                curr_area = (brx - blx) * (tlyy - bly)
                                min_area = min(min_area, curr_area)
                                    
        if min_area == float('inf'):
            return 0
        return min_area

    
    def group_by_x(self, points):
        res = defaultdict(list)
        
        for pt in points:
            res[pt[0]].append(pt)
        return res
    
    def group_by_y(self, points):
        res = defaultdict(list)
        
        for pt in points:
            res[pt[1]].append(pt)
        return res
```

# Accepted, using set to find matching top right corner

```python
from collections import defaultdict
class Solution:
    def minAreaRect(self, points: List[List[int]]) -> int:
        y_groups = self.group_by_y(points)
        x_groups = self.group_by_x(points)
        memo = set()
        for point in points:
            memo.add((point[0],point[1]))
        #print(y_groups)
        #print(x_groups)
        #print()
        min_area = float('inf')
        for lvl in y_groups:
            pt_list = y_groups[lvl]
            pt_list.sort()
            n = len(pt_list)
            for i in range(n):
                blx,bly = pt_list[i]
                #print(pt_list[i])
                for j in range(i + 1, n):
                    brx,bry = pt_list[j]
                    #print(pt_list[j])
                    tl_cands = x_groups[blx]
                    for tl_cand in tl_cands:
                        tlx, tlyy = tl_cand
                        if tlyy > bly and (brx, tlyy) in memo:
                            curr_area = (brx - blx) * (tlyy - bly)
                            min_area = min(min_area, curr_area)
                 
        if min_area == float('inf'):
            return 0
        return min_area

    
    def group_by_x(self, points):
        res = defaultdict(list)
        
        for pt in points:
            res[pt[0]].append(pt)
        return res
    
    def group_by_y(self, points):
        res = defaultdict(list)
        
        for pt in points:
            res[pt[1]].append(pt)
        return res
```
