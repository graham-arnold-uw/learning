```python
import copy
class NumMatrix:

    def __init__(self, matrix: List[List[int]]):
        self.matrix = copy.deepcopy(matrix)
        self.m = len(self.matrix)
        self.n = len(self.matrix[0])
        self.summat = self.initialize_sum_mat()
      #  for i in range(self.m):
       #     print(self.matrix[i])
        #print()
        #for i in range(self.m):
        #    print(self.summat[i])
        
        
    def initialize_sum_mat(self):
        res = [[0 for j in range(self.n)] for i in range(self.m)]
        
        for col in range(self.n - 1, -1, -1):
            if col == self.n - 1:
                res[self.m - 1][col] = self.matrix[self.m - 1][col]
            else:
                res[self.m - 1][col] = res[self.m - 1][col + 1] + self.matrix[self.m - 1][col]
        
        for row in range(self.m - 2, -1, -1):
            #if row == self.m - 1:
                #res[row][self.n - 1] = matrix[row][self.n - 1]
            #else:
            res[row][self.n - 1] = res[row + 1][self.n - 1] + self.matrix[row][self.n - 1]
                
        for row in range(self.m - 2, -1, -1):
            for col in range(self.n - 2, -1, -1):
                res[row][col] = res[row + 1][col] + res[row][col + 1] - res[row + 1][col + 1] + self.matrix[row][col]
        return res

    def update(self, row: int, col: int, val: int) -> None:
        #for i in range(self.m):
        #    print(self.summat[i])
        #print()
        old_val = self.matrix[row][col]
        self.matrix[row][col] = val
        delta = val - old_val
        
        for r in range(row, -1, -1):
            for c in range(col, -1, -1):
                self.summat[r][c] += delta
                
                
       # for i in range(self.m):
        #    print(self.summat[i])
        #print()

    def sumRegion(self, row1: int, col1: int, row2: int, col2: int) -> int:
        h = row2 - row1 + 1
        w = col2 - col1 + 1
        
        trr, trc = row1, col1 + w
        blr, blc = row1 + h, col1
        brr, brc = row1 + h, col1 + w
        
        bl_cont = tr_cont = br_cont = 0
        if blr >= 0 and blr < self.m and blc >= 0 and blc < self.n:
            bl_cont = self.summat[blr][blc]
        
        if trr >= 0 and trr < self.m and trc >= 0 and trc < self.n:
            tr_cont = self.summat[trr][trc]
            
        if tr_cont != 0 and bl_cont != 0:
             #if brr >= 0 and brr < self.m and brc >= 0 and brc < self.n:
            br_cont = self.summat[brr][brc]
            
        #print(tr_cont)
        #print(bl_cont)
        #print(br_cont)
        #print()
        tot = self.summat[row1][col1] - bl_cont - tr_cont + br_cont
        return tot


# Your NumMatrix object will be instantiated and called as such:
# obj = NumMatrix(matrix)
# obj.update(row,col,val)
# param_2 = obj.sumRegion(row1,col1,row2,col2)
```
