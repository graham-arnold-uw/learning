```python
class Solution:
    def findMedianSortedArrays(self, nums1: List[int], nums2: List[int]) -> float:
        n1 = len(nums1)
        n2 = len(nums2)
        
        if n1 == 0:
            if n2 % 2 == 0:
                m1 = nums2[n2 // 2 - 1]
                m2 = nums2[n2 // 2]
                return (m1 + m2) / 2
            else:
                return nums2[n2 // 2]
        elif n2 == 0:
            if n1 % 2 == 0:
                m1 = nums1[n1 // 2 - 1]
                m2 = nums1[n1 // 2]
                return (m1 + m2) / 2
            else:
                return nums1[n1 // 2]
        
        total = n1 + n2
        odd = total % 2 == 1
        
        s1, e1 = 0, n1 - 1
        s2, e2 = 0, n2 - 1
    
        while s1 <= e1 and s2 <= e2:
            if not odd and s1 == e1 and s2 == e2:
                return (nums1[s1] + nums2[s2]) / 2
            
            if nums1[s1] <= nums2[s2]: 
                s1 += 1
            else:
                #print(f"{s1} {e1} {e2}")
                s2 += 1
            
            if nums1[e1] >= nums2[e2]:
                e1 -= 1
            else:
                e2 -= 1
        
        #print(f"s1 = {s1} e1 = {e1}")
        #print(f"s2 = {s2} e2 = {e2}")
        #print()
        if odd:
            if e1 >= s1:
                #get median 
                while e1 > s1:
                    s1 += 1
                    e1 -= 1
                return nums1[s1]
                
            else:
                while e2 > s2:
                    s2 += 1
                    e2 -= 1
                #median is at s2
                return nums2[s2]
            
        else:
            if e1 == s1 and e2 == s2:
                return (nums1[e1] + nums2[e2]) / 2
            
            elif e1 > s1:
                while e1 >= s1 + 1:
                    s1 += 1
                    e1 -= 1
                return (nums1[e1] + nums1[s1]) / 2
            else: # e2 > s2
                while e2 >= s2 + 1:
                    #print('h2')
                    s2 += 1
                    e2 -= 1
                #print(f"s2 = {s2} e2 = {e2}" )
                return (nums2[e2] + nums2[s2]) / 2   
                
                
                
                
                
                
```
